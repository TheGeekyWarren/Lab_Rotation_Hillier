```python
import cv2
import os

def conv_vids(folder_path, name):

    # specify the desired frame rate of the output video
    fps = 10
    print('start')
    namev = name + '.avi'
    # get a list of all the image file names in the folder
    image_files = []
    for f in os.listdir(folder_path):
        if f.endswith(".png"):
            image_files.append(f)
    #print(image_files)

    # sort the list of image file names in ascending order
    image_files.sort()

    # get the dimensions of the first image in the folder
    img = cv2.imread(os.path.join(folder_path, image_files[0]))
    height, width, channels = img.shape

    # create a VideoWriter object to write the output video
    fourcc = cv2.VideoWriter_fourcc(*"XVID")
    video_writer = cv2.VideoWriter(namev, fourcc, fps, (width, height))

    font = cv2.FONT_HERSHEY_DUPLEX
    color = (255, 255, 255) # red
    fontsize = 0.45
    position = (20, 20)
    
 
    # open a text file to write the text
    text_file = open("text_file.txt", "w")

    # loop over all the image files and add them to the output video
    for image_file in image_files:
        #read the image file
        img = cv2.imread(os.path.join(folder_path, image_file))

        # put the text on the image
        text = "Frame: " + image_file + "\n"
        cv2.putText(img, text, position, font, fontsize, color=color)
    
        # write the text to the text file
        text_file.write(text)

        # write the image to the output video
        video_writer.write(img)
    print('done')

    # close the text file
    text_file.close()
     # release the video writer and destroy any remaining windows
    video_writer.release()
    cv2.destroyAllWindows()
```
