```python
import time
import random
import pygame
# initialize pygame
pygame.init()

# define width of screen
width = 1600
# define height of screen
height = 900
screen_res = (width, height)

pygame.display.set_caption("GFG Bouncing game")
screen = pygame.display.set_mode(screen_res)
correct_clicks = 0

for trial in range(1, 2): #number of trials --------------------------------------------------------------------------------------------------------------------------------------------
    rb = 40 #radius of the ball---------------------------------------------------------------------------------------------------------------------------------------------------------
    x = random.randint(100, width-100)
    y = random.randint(100, height-100)
    # Define the fixation point coordinates
    fixation = [x,y]
    fixation_x, fixation_y = fixation #randomize this
    fixation_radius = rb

    # Draw the fixation point
    fixation_color = (255, 255, 255)
    pygame.draw.circle(screen, fixation_color, (fixation_x, fixation_y), fixation_radius)

    # Update the display
    pygame.display.flip()

    # Wait for a mouse click on the fixation point to start the trial
    trial_started = False
    while not trial_started:
        for event in pygame.event.get():
            if event.type == pygame.MOUSEBUTTONDOWN:
                mouse_x, mouse_y = event.pos
                if (mouse_x - fixation_x)**2 + (mouse_y - fixation_y)**2 <= fixation_radius**2:
                    trial_started = True


    # define colors
    red = (255, 0, 0)
    dred = (200, 0, 0)
    black = (0, 0, 0)

    # define ball
    ball_obj = pygame.draw.circle(surface=screen, color=red, center=fixation, radius=rb)
    
    # define speed of ball
    
    # speed = [X direction speed, Y direction speed]
    
    speed = [1, 1] #-------------------------------------------------------------------------------------------------------------------------------------------------------------------

    limit = random.randint(3, 7) #time for which ball is seen--------------------------------------------------------------------------------------------------------------------------

    last_seen = []
    all_points = []

    start_time = time.time()
    
    flickr = random.randint(1,limit-1)
    bound = [start_time + flickr, start_time+flickr+0.15]
    count = 0

    # game loop
    while True:
        # event loop

        if (time.time() - start_time) >= limit:
            last_seen = ball_obj.center
            break

        # fill black color on screen
        screen.fill(black)

        # move the ball
        # Let center of the ball is (100,100) and the speed is (1,1)
        ball_obj = ball_obj.move(speed)
        # Now center of the ball is (101,101)
        # In this way our wall will move

        # if ball goes out of screen then change direction of movement
        if ball_obj.left <= 0 or ball_obj.right >= width:
            speed[0] = -speed[0]
        if ball_obj.top <= 0 or ball_obj.bottom >= height:
            speed[1] = -speed[1]
        
        if time.time() >= bound[0] and time.time() <= bound[1]:
            pygame.draw.circle(surface=screen, color=dred, center=ball_obj.center, radius=rb)  
            print(time.time())
        # draw ball at new centers that are obtained after moving ball_obj
        else:
            pygame.draw.circle(surface=screen, color=red, center=ball_obj.center, radius=rb)

        # update screen
        pygame.display.flip()

    for i in range(1, 1000):
        pygame.draw.circle(surface=screen, color=black,
                        center=ball_obj.center, radius=rb)
        pygame.display.flip()

    # for i in range(1, 1000):
    #     screen.fill((0,0,0))

    end_x, end_y = last_seen
    end_rad = rb*5 #error margin for correct click----------------------------------------------------------------------------------------------------------------------------------------
    
    click = False
    while not click:
        for event in pygame.event.get():
            if event.type == pygame.MOUSEBUTTONDOWN:
                mc = event.pos
                mouse_x, mouse_y = mc
                if (mouse_x - end_x)**2 + (mouse_y - end_y)**2 <= end_rad**2:
                    correct_clicks = correct_clicks + 1
                click = True
    print(last_seen)
    print(mc)
    print("")
                    
for i in range(1, 1000): #displays results for a few seconds
    font = pygame.font.SysFont(None, 48)
    text = font.render(f"Correct clicks = {correct_clicks}", True, (255, 255, 255))
    screen.blit(text, (width/2 - text.get_width()/2, height/2 - text.get_height()/2))
    pygame.display.flip()

pygame.display.quit()
exit()
```
